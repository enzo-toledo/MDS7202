### Extra: ¿Qué es el CI y qué verifican estas herramientas?

**CI (Integración Continua)** es una práctica en la que cada vez que se sube código al repositorio (mediante un push o un Pull Request), se ejecutan automáticamente una serie de verificaciones para asegurar que el código cumple ciertos estándares de calidad. En este curso, el CI está configurado en `.github/workflows/ci.yml` y es ejecutado por GitHub Actions.

Las herramientas que configuramos cumplen roles complementarios:

| Herramienta | ¿Qué verifica? |
|---|---|
| **`ruff` (linter)** | Detecta errores de estilo, imports no usados, variables indefinidas, patrones propensos a bugs y sintaxis desactualizada |
| **`ruff` (formatter)** | Formatea el código automáticamente para que tenga un estilo consistente (espacios, saltos de línea, comillas, etc.) |
| **`pre-commit`** | Ejecuta un conjunto de revisiones automáticamente *antes de cada commit*, impidiendo subir código que no pase las verificaciones |

En la práctica: cuando hagan `git commit`, `pre-commit` correrá `ruff` sobre sus archivos. Si hay problemas, el commit será bloqueado y deberán corregirlos antes de continuar. Esto asegura que todo el código que llegue al repositorio sea consistente y libre de errores básicos.

#### Pregunta 1 (0.5 pts)

Dada una lista de $n$ elementos, no necesariamente ordenada, defina funciones que busquen:
- El elemento menor.
- El elemento mayor.
- El promedio de los elementos en la lista.

Ejemplo:
```python
lista_n = [12,4,34,1,2,52,1,54,98,5,235,76,57,87,123,32,32]
menor = 1
mayor = 235
```


````python
def maximo(lista):
    if lista == []:
        return None
    elif len(lista) == 1:
        return lista[0]
    else:
        max_elem = lista[0]
        for elem in lista:
            if elem > max_elem:
                max_elem = elem
        return max_elem
````

````python
lista1 = [3, 1, 4, 1, 5, 9]
lista2 = []
lista3 = [42]
lista4 = [12, 4, 34, 1, 2, 52, 1, 54, 98, 5, 235, 76, 57, 87, 123, 32, 32]

print(maximo(lista1))
print(maximo(lista2))
print(maximo(lista3))
print(maximo(lista4))
````

````python
def minimo(lista):
    if lista == []:
        return None
    elif len(lista) == 1:
        return lista[0]
    else:
        min_elem = lista[0]
        for elem in lista:
            if elem < min_elem:
                min_elem = elem
        return min_elem
````

````python
print(minimo(lista1))
print(minimo(lista2))
print(minimo(lista3))
print(minimo(lista4))
````

````python
def promedio(lista):
    if lista == []:
        return None
    else:
        suma = 0
        for elem in lista:
            suma += elem
        return suma / len(lista)
````

````python
print(promedio(lista1))
print(promedio(lista2))
print(promedio(lista3))
````

#### Pregunta 2 (0.5 pts)
Genere una función que obtenga la moda de una lista usando diccionarios.

````python
def moda(lista):
    if lista == []:
        return None
    else:
        frecuencia = {}
        for elem in lista:
            if elem in frecuencia:
                frecuencia[elem] += 1
            else:
                frecuencia[elem] = 1
        moda_elem = None
        max_frecuencia = 0
        for elem, freq in frecuencia.items():
            if freq > max_frecuencia:
                max_frecuencia = freq
                moda_elem = elem
        return moda_elem
````

#### Pregunta 3 (0.5 pts)
Vea si un número es palíndromo o no. Los números palíndromos son aquellos que al leerlos al revés y al derecho son iguales, por ejemplo el 123321 o el 98789.

````python
def es_palindromo(num):
    if num < 0:
        return False
    else:
        s_num = str(num)
        inv_s = str(num)[::-1]
        return s_num == inv_s
````

#### Pregunta 4 (0.5 pts)
Dado un texto, conviértalo en estilo Camel Case, vea el siguiente ejemplo:

    Texto inicial: “Este es un texto de ejemplo para mostrar el Camel Case”.

    Texto final: “EsteEsUnTextoDeEjemploParaMostrarElCamelCase”.

````python
def to_cc(string):
    # Separamos palabras de la oracion
    x = string.split(sep=" ")
    y = []
    for elem in x:
        if elem != "":
            y.append(elem)

    # Escribimos la primera letra de cada palabra en mayuscula
    z = []
    for elem in y:
        first_letter = elem[0].upper()
        rest = elem[1:]
        z.append(first_letter + rest)

    # Guardamos y entregamos la suma de las palabras
    result = ""
    for elem in z:
        result += elem

    return result
````

````python
to_cc("Este es un texto de ejemplo para mostrar el Camel Case")
````

````python
to_cc("Este es  un texto de    ejemplo para mostrar   el  Camel Case")
````

#### Pregunta 1 (0.5 pts):

Un monto inicial de dinero (por ejemplo un millón de pesos) se puede depositar en un banco con un interés anual de 5%. Esto significa  que  cumplido  un  año  se  dispondrá  de  \\$1.050.000. Si se deposita por dos años, entonces  al  finalizar  el  2º  año se dispondrá de \\$1.102.500,  es  decir,  \\$1.050.000  reajustado  en  5%.  

Al  respecto, si  se  invoca  a  la  función  `calc_capital_final(1000000, 5, 2)`  entrega  el resultado `1050000.0`. 

Escriba la función `calc_capital_final`

$$C_f = C_i(1+i)^n$$

- $C_f$ = Capital Final
- $C_i$ = Capital Inicial
- $i$ = Tasa de interés
- $n$ = Periodo del ahorro

Además, incluya tests que certifiquen el correcto funcionamiento de la función.

````python
def calc_capital_final(capital_inicial, tasa_interes, anios):
    # Posibles errores para el capital inicial
    if not isinstance(capital_inicial, int | float):
        raise TypeError("El capital inicial debe ser un numero")
    if capital_inicial < 0:
        raise ValueError("El capital inicial no puede ser negativo")

    # Posibles errores para la cantidad de anios
    if not isinstance(anios, int):
        raise TypeError("La cantidad de anios debe ser un entero")
    if anios < 0:
        raise ValueError("La cantidad de anios debe ser positiva")

    # Posibles errores para la tasa de interes
    if not isinstance(tasa_interes, int | float):
        raise TypeError("La tasa de interes debe ser un numero")
    if (0 > tasa_interes) or (tasa_interes > 100):
        raise ValueError("La tasa de interes debe estar entre 0 y 100")

    # Formula
    c_f = capital_inicial * (1 + tasa_interes / 100) ** anios

    return c_f
````

````python
def tests():
    """
    Testeos de la función calc_capital_final.
    """
    assert calc_capital_final(1000000, 5, 1) == 1050000.0
    assert calc_capital_final(1000000, 5, 2) == 1102500.0
    assert calc_capital_final(1000000, 5, 0) == 1000000.0
    assert calc_capital_final(1000000, 0, 0) == 1000000.0
    assert calc_capital_final(0, 0, 0) == 0.0
````

````python
tests()
````

#### Pregunta 2 (0.5 pts):

Un anagrama es una palabra que resulta de la transposición de letras de otra palabra o frase.

Por ejemplo, las palabras “caso” y “caos” sonan agramas de “saco”  y las palabras “acaso” y “ocaso” no son anagramas de “caso”. Las anagramas poseen las mismas letras, con la misma cantidad de apariciones, pero en un orden diferente

Escriba la función `def anagrama(x,y)` que reciba dos palabras y entregue `True` si son anagramas y `False` si no.


````python
def anagrama(x, y):
    # Si x e y son palabras de diferentes largos, entonces no son anagramas
    if len(x) != len(y):
        return False

    # Recorremos x, y anotamos la frecuencia de sus letras
    let_x = []
    for letter in x:
        let_x.append(letter)

    frec_x = {}
    for elem in let_x:
        if elem in frec_x:
            frec_x[elem] += 1
        else:
            frec_x[elem] = 1

    # Recorremos y, y anotamos la frecuencia de sus letras
    let_y = []
    for letter in y:
        let_y.append(letter)

    frec_y = {}
    for elem in let_y:
        if elem in frec_y:
            frec_y[elem] += 1
        else:
            frec_y[elem] = 1

    # Revisamos que las letras de x esten en y, y al reves
    for let in let_x:
        if let not in let_y:
            return False
    for let in let_y:
        if let not in let_x:
            return False

    # Revisamos que se repitan la misma cantidad de veces en cada palabra
    for let_x, times_x in frec_x.items():
        if frec_y[let_x] != times_x:
            return False

    for let_y, times_y in frec_y.items():
        if frec_x[let_y] != times_y:
            return False

    return True
````

````python
def tests_anagrama():
    """
    Testeos de la función anagrama.
    """
    assert anagrama("caso", "caos")
    assert not anagrama("acaso", "ocaso")
    assert anagrama("7890.3", "3089.7")
    assert not anagrama(" ", "")
    assert anagrama("", "")
````

````python
tests_anagrama()
````
